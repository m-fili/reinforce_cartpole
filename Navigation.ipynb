```python
import numpy as np
import pandas as pd
import gym
import matplotlib.pyplot as plt
from reinforce_algorithm import Agent, MLP, reinforce
```

```python
env = gym.make('CartPole-v1')
model = MLP(n_input=4, n_output=2, n_hidden=[16])
agent = Agent(env, model)

scores = reinforce(agent, n_episodes=1000, m=1, gamma=1.0, t_max=1000, learning_rate=5e-3)
```

### Visualization

```python
plt.figure(figsize=(8, 3.5))

plt.plot(scores, alpha=0.2, color='purple', label='scores')
plt.plot(pd.Series(scores).rolling(100).mean(), color='red', label='average score')

plt.xlabel('Episode')
plt.ylabel('Score')
plt.legend()
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
```
